```python
#format date
covid_df['Date']=pd.to_datetime(covid_df['Date'],format='%Y-%m-%d')
```

```python
#Creation Pivot Tables Very important
statewise=pd.pivot_table(covid_df,values=['Confirmed','Cured','Deaths'],index='place',aggfunc=max)
statewise.head()
```

```python
statewise['Recovery_Rate']=(statewise['Cured']*100/statewise['Confirmed'])
statewise['Mortality_Rate']=(statewise['Deaths']*100/statewise['Confirmed'])
statewise.head()
```

```python
statewise=statewise.sort_values(by='Confirmed', ascending=False)
statewise.head()
```

```python

#styling a pivot table very interesting
statewise.style.background_gradient(cmap='cubehelix')
```

```python
#to calculated active cases
covid_df['Active_case']=covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
```

```python
#Top 10 active cases
top_10_active_cases=covid_df.groupby(by='place').max()[['Active_case','Date']]
```
